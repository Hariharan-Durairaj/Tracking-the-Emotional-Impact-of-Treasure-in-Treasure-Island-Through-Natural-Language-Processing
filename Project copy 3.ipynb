```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

def plot_emotion_heatmap(results, file_name=""):
    # Convert results to DataFrame
    data = []
    for chap, values in results.items():
        for emotion, score in values.items():
            data.append({'Chapter': chap, 'Emotion': emotion.capitalize(), 'Score': score})
    df = pd.DataFrame(data)

    # Sort emotions
    emotion_order = ['Fear', 'Excitement', 'Greed', 'Danger', 'Joy', 'Surprise']
    df['Emotion'] = pd.Categorical(df['Emotion'], categories=emotion_order, ordered=True)

    # Pivot for heatmap
    pivot_df = df.pivot(index="Chapter", columns="Emotion", values="Score").sort_index()

    # Plot with figure and axis for full control
    fig, ax = plt.subplots(figsize=(8, 9))
    sns.set(style="whitegrid", font_scale=1.1)
    
    sns.heatmap(pivot_df, annot=True, cmap="YlOrRd", fmt=".2f",
                linewidths=0.5, linecolor='gray',
                cbar_kws={'label': 'Emotion Score'}, ax=ax)

    # Title and labels
    fig.suptitle("Heatmap of Emotion Intensity Across Chapters",
                 fontsize=23, fontweight='bold', y=1.02)

    ax.set_xlabel("Emotion", fontsize=20)
    ax.set_ylabel("Chapter", fontsize=20)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=18)
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=16)

    fig.tight_layout(rect=[0, 0, 1, 1.03])  # reserve space for title
    if file_name!="":
        fig.savefig(f"emotion_heatmap_{file_name}.png", dpi=300)
    plt.show()

def calculate_emotion_scores(emotion_context_scores, aggregation_method, lambda_decay=0.1):
    emotion_scores = {}
    for emotion, scores in emotion_context_scores.items():
        if not scores:
            emotion_scores[emotion] = 0
            continue

        sims = [sim for sim, dist in scores]

        if aggregation_method == "average":
            emotion_scores[emotion] = np.mean(sims)            
        elif aggregation_method == "max":
            emotion_scores[emotion] = np.max(sims)
        elif aggregation_method == "median":
            emotion_scores[emotion] = np.median(sims)
        elif aggregation_method == "weighted_average":
            weighted_scores = []
            weights = []
            for sim, dist in scores:
                weight = 1 / (dist + 1)
                weighted_scores.append(sim * weight)
                weights.append(weight)
            emotion_scores[emotion] = sum(weighted_scores) / sum(weights)
        elif aggregation_method == "exp_decay":
            weighted_scores = []
            weights = []
            for sim, dist in scores:
                weight = np.exp(-lambda_decay * dist)
                weighted_scores.append(sim * weight)
                weights.append(weight)
            emotion_scores[emotion] = sum(weighted_scores) / sum(weights)
        else:
            raise ValueError(f"Unknown aggregation method: {aggregation_method}")
    return emotion_scores


def plot_emotion_heatmap_normalized(results, file_name="", plot=True):
    # Convert results to DataFrame
    data = []
    for chap, values in results.items():
        for emotion, score in values.items():
            data.append({'Chapter': chap, 'Emotion': emotion.capitalize(), 'Score': score})
    df = pd.DataFrame(data)

    # Sort emotions
    emotion_order = ['Fear', 'Excitement', 'Greed', 'Danger', 'Joy', 'Surprise']
    df['Emotion'] = pd.Categorical(df['Emotion'], categories=emotion_order, ordered=True)

    # Pivot for heatmap
    pivot_df = df.pivot(index="Chapter", columns="Emotion", values="Score").sort_index()

    scaler = StandardScaler()
    df_normalized = pd.DataFrame(scaler.fit_transform(pivot_df), columns=pivot_df.columns, index=pivot_df.index)
    df_clipped = df_normalized.clip(lower=-1, upper=1)
    result_normalized = (df_clipped +1) /2
    # Plot with figure and axis for full control
    if plot:
        fig, ax = plt.subplots(figsize=(8, 9))
        sns.set(style="whitegrid", font_scale=1.1)

        sns.heatmap(result_normalized, annot=True, cmap="YlOrRd", fmt=".2f",
                    linewidths=0.5, linecolor='gray',
                    cbar_kws={'label': 'Emotion Score'}, ax=ax)

        # Title and labels
        fig.suptitle("Heatmap of Emotion Intensity Across Chapters",
                    fontsize=23, fontweight='bold', y=1.02)

        ax.set_xlabel("Emotion", fontsize=20)
        ax.set_ylabel("Chapter", fontsize=20)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=18)
        ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=16)

        fig.tight_layout(rect=[0, 0, 1, 1.03])  # reserve space for title
        if file_name != "":
            fig.savefig(f"emotion_heatmap_{file_name}.png", dpi=300)
        plt.show()
    #result_normalized = (df_clipped +1) /2
    result_normalized= result_normalized.reset_index()
    result_normalized = pd.DataFrame(result_normalized.reset_index(), columns=['Chapter', 'Fear', 'Excitement', 'Greed', 'Danger', 'Joy', 'Surprise'])
    return result_normalized 

from sklearn.metrics import roc_auc_score
def calculate_roc_auc(result_normalized, chapter_emotion_scores=chapter_emotion_scores):
    sum=[]
    for emotion in result_normalized.columns[1:]:
        y_true = [chapter_emotion_scores[i][emotion] for i in result_normalized.Chapter.values]
        y_score = result_normalized[emotion].values
        auc = roc_auc_score(y_true, y_score)
        sum.append(round(auc,3))
        print(f"{emotion} AUC: {auc:.3f}")
    print()
    print(f"Average AUC:{ np.mean(sum):.3f}")
    return sum, np.mean(sum)

from sklearn.metrics import f1_score
def calculate_f1_score(result_normalized, threshold=0.5, chapter_emotion_scores=chapter_emotion_scores):
    sum=[]
    for emotion in result_normalized.columns[1:]:
        y_true = [chapter_emotion_scores[i][emotion] for i in result_normalized.Chapter.values]
        y_pred = (result_normalized[emotion].values > threshold).astype(int)
        f1 = f1_score(y_true, y_pred)
        sum.append(round(f1,3))
        print(f"{emotion} F1 Score: {f1:.3f}")
    print()
    print(f"Average F1 Score:{ np.mean(sum):.3f}")
    return sum, np.mean(sum)

from sklearn.calibration import calibration_curve
def calculate_calibration_curve(result_normalized, chapter_emotion_scores=chapter_emotion_scores):
    emotions = result_normalized.columns[1:]
    plt.figure(figsize=(14, 10))
    for i, emotion in enumerate(emotions, 1):
        plt.subplot(2, 3, i)
        prob_true, prob_pred = calibration_curve([chapter_emotion_scores[i][emotion] for i in result_normalized.Chapter.values], result_normalized[emotion], n_bins=10)
        plt.plot(prob_pred, prob_true, marker='o')
        plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
        plt.title(f'Calibration Curve: {emotion}')
        plt.xlabel('Mean Predicted Probability')
        plt.ylabel('True Probability')
        plt.grid()

    plt.tight_layout()
    plt.show()

from sklearn.metrics import roc_curve, auc
def calculate_roc_curve(result_normalized, chapter_emotion_scores=chapter_emotion_scores):
    manual_scores_df = pd.DataFrame.from_dict(chapter_emotion_scores, orient='index')
    manual_scores_df.reset_index(drop=True, inplace=True)

    # List of emotions
    emotions = ["Fear", "Excitement", "Greed", "Danger", "Joy", "Surprise"]

    # Plot ROC for each emotion
    plt.figure(figsize=(12, 8))

    for emotion in emotions:
        y_true = manual_scores_df[emotion]
        y_scores = result_normalized[emotion]
        
        fpr, tpr, _ = roc_curve(y_true, y_scores)
        roc_auc = auc(fpr, tpr)

        plt.plot(fpr, tpr, label=f"{emotion} (AUC = {roc_auc:.2f})")

    plt.plot([0, 1], [0, 1], 'k--')  # diagonal line
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves for Emotions")
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_emotion_similarity_histograms(emotion_context_scores, figsize=(8,9), file_name=""):
    # Plot setup
    plt.figure(figsize=figsize)
    plt.suptitle('Histograms of Emotion Similarity Scores', fontsize=23, fontweight='bold', y=1.05)
    
    for idx, (emotion, scores) in enumerate(emotion_context_scores.items(), 1):
        plt.subplot(2, 3, idx)  # 2 rows, 3 columns
        plt.hist(scores, bins=10, color='skyblue', edgecolor='black')
        plt.title(emotion.capitalize(), fontsize=18)
        plt.xlabel('Similarity Score', fontsize=13)
        plt.ylabel('Frequency', fontsize=13)
        plt.xlim(0, 1)
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)  # So title doesn't overlap

    # Save the figure
    if file_name != "":
        plt.savefig(f"emotion_similarity_histograms_{file_name}.png", dpi=400)
    plt.show()

import pandas as pd
from itertools import product

def run_emotion_analysis_pipeline(
    
    aggregation_methods=["weighted_average", "average","median", "max", "exp_decay"],
    window_sizes=[5, 10, 15],
    lambda_decays=[0.3, 0.1,0.06],
    f1_thresholds=[0.40, 0.50, 0.60],
    similarity_metrics=["cosine", "euclidean", "pearson"],
    nlp_models=None,
    remove_stopwords=True,
    alpha_values=[0.25,0.50,0.75]

):
    """
    Runs full emotion analysis with gridsearch over aggregation_method, window_size,
    and supports multiple NLP models.
    
    Parameters:
        book: str - raw text of the book
        EMOTION_SEEDS: dict - seeds for emotion detection
        aggregation_methods: list - list of aggregation methods to test
        window_sizes: list - list of window sizes to test
        f1_thresholds: list - thresholds for F1 calculation
        nlp_models: dict - {"model_name": analyze_emotions_function} (optional)
    
    Returns:
        results_df: DataFrame - combined results of AUC and F1
    """

    # Step 1: Clean text by chapter (one time)
    chapters = clean_text_by_chapter(book, remove_stopwords=remove_stopwords)

    results_list = []

    # If no NLP models given, default to analyze_emotions_word2vec
    

    # Step 2: Loop through models, aggregation_methods, and window_sizes
    for model_name, analyze_func in nlp_models.items():
        for sim_metric in similarity_metrics:
            
            for window_size in window_sizes:
                
                for alpha in alpha_values:
                    if sim_metric == "pearson" and analyze_func.__name__.lower().__contains__("spacy"):
                        continue  # Skip Pearson for spaCy models
                    results, all_emotion_context_scores = analyze_func(
                        chapters,
                        EMOTION_SEEDS,
                        window_size = window_size,
                        similarity_aggregation = "max",
                        similarity_metric = sim_metric,
                        lambda_decay=0.1,
                        alpha=alpha
                    )
                    
                    for agg_method in aggregation_methods:
                        
                        
                        if agg_method != "exp_decay":
                            
                            for chap_num,emotion_context_scores in all_emotion_context_scores.items():
                                
                                results[chap_num] = calculate_emotion_scores(emotion_context_scores[0], agg_method)
                            # Step 4: Plot heatmap and get normalized results
                            
                            result_normalized = plot_emotion_heatmap_normalized(results, plot=False)

                            # Step 5: Calculate ROC AUC
                            auc_scores, auc_macro = calculate_roc_auc(result_normalized)

                            for f1_threshold in f1_thresholds:
                                # Step 6: Calculate F1
                                f1_scores, f1_macro = calculate_f1_score(result_normalized, threshold=f1_threshold)

                                # Step 7: Store results in a list of dicts
                                results_list.append({
                                    "Model": model_name,
                                    "Similarity _Metric": sim_metric,
                                    "Aggregation_Method": agg_method,
                                    "lambda_Decay": None,  # Not applicable for non-exp_decay methods
                                    "alpha": alpha if analyze_func.__name__.lower().__contains__("vad") else None,
                                    "Window_Size": window_size,
                                    "F1_Threshold": f1_threshold,
                                    "AUC_Scores": auc_scores,
                                    "AUC_Macro": auc_macro,
                                    "F1_Scores": f1_scores,
                                    "F1_Macro": f1_macro
                                })

                        elif agg_method == "exp_decay":
                            
                            for lamda_decay in lambda_decays:
                                for chap_num,emotion_context_scores in all_emotion_context_scores.items():
                                    results[chap_num] = calculate_emotion_scores(emotion_context_scores[0], agg_method, lamda_decay)
                                # Step 4: Plot heatmap and get normalized results
                                
                                result_normalized = plot_emotion_heatmap_normalized(results, plot=False)

                                # Step 5: Calculate ROC AUC
                                auc_scores, auc_macro = calculate_roc_auc(result_normalized)

                                for f1_threshold in f1_thresholds:
                                    # Step 6: Calculate F1
                                    f1_scores, f1_macro = calculate_f1_score(result_normalized, threshold=f1_threshold)

                                    # Step 7: Store results in a list of dicts
                                    results_list.append({
                                        "Model": model_name,
                                        "Similarity _Metric": sim_metric,
                                        "Aggregation_Method": agg_method,
                                        "lambda_Decay": lamda_decay,
                                        "alpha": alpha if analyze_func.__name__.lower().__contains__("vad") else None,
                                        "Window_Size": window_size,
                                        "F1_Threshold": f1_threshold,
                                        "AUC_Scores": auc_scores,
                                        "AUC_Macro": auc_macro,
                                        "F1_Scores": f1_scores,
                                        "F1_Macro": f1_macro
                                    })
                    if not analyze_func.__name__.lower().__contains__("vad"):
                        break
                            
            # Step 8: Convert to DataFrame
    results_df = pd.DataFrame(results_list)

    return results_df

from scipy.spatial.distance import cosine, euclidean
from scipy.stats import pearsonr
def vector_similarity(vec1, vec2, metric="pearson"):
    
    if metric == "pearson":
        return pearsonr(vec1, vec2)[0]  # Pearson correlation coefficient
    elif metric == "euclidean":
        # Convert distance to similarity (smaller distance → higher similarity)
        dist = euclidean(vec1, vec2)
        return 1 / (1 + dist)
    else:
        raise ValueError(f"Unknown similarity metric: {metric}")


def get_vad_vector(word, vad_df):
    """Return VAD vector for word or None if missing."""
    row = vad_df[vad_df['word'] == word]
    if not row.empty:
        return row[['Valence', 'Arousal', 'Dominance']].values[0]
    return None

def vad_similarity(vec1, vec2):
    """Convert Euclidean distance between VAD vectors into similarity [0,1]."""
    dist = np.linalg.norm(vec1 - vec2)
    return 1 / (1 + dist)  # inverse distance


# -------------------------
# 2. Precompute emotion VADs
# -------------------------
def compute_emotion_vad_profiles(emotion_seeds, vad_df):
    """Compute the average VAD vector for each emotion seed list."""
    profiles = {}
    for emotion, words in emotion_seeds.items():
        vad_vectors = [get_vad_vector(w, vad_df) for w in words]
        vad_vectors = [v for v in vad_vectors if v is not None]
        if vad_vectors:
            profiles[emotion] = np.mean(vad_vectors, axis=0)
        else:
            profiles[emotion] = None
    return profiles
```

```python
import numpy as np
import pandas as pd
import spacy
from collections import defaultdict

nlp = spacy.load("en_core_web_lg")

def analyze_emotions_spacy_with_vad(chapters, EMOTION_SEEDS, window_size=10, aggregation_method="average", lambda_decay=0.5, similarity_aggregation="max", similarity_metric="cosine" ,alpha=0.75, **kwargs):
    # Precompute emotion seed docs for embeddings
    emotion_docs = {
        emotion: nlp(" ".join(words))
        for emotion, words in EMOTION_SEEDS.items()
    }

    # Precompute emotion VAD profiles
    emotion_vad_profiles = compute_emotion_vad_profiles(EMOTION_SEEDS, VAD)

    results = {}
    all_emotion_context_scores = defaultdict(list)

    print(f"Processing Chapter ", end="")
    for chap_num, text in chapters.items():
        print(f"{chap_num}, ", end="")
        doc = nlp(text)
        emotion_context_scores = defaultdict(list)

        for i, token in enumerate(doc):
            if token.lemma_.lower() in ["treasure", "gold", "map", "rich", "island", "ship", "chart", "x"]:
                start = max(0, i - window_size)
                end = min(len(doc), i + window_size + 1)
                context = doc[start:end]

                for j, ctx in enumerate(context):
                    if not ctx.is_stop and not ctx.is_punct and ctx.has_vector:
                        distance = abs(j - (i - start))

                        for emotion, emotion_doc in emotion_docs.items():
                            if similarity_metric == "cosine":
                                # Use spaCy's built-in similarity
                                similarities = [ctx.similarity(seed_word) for seed_word in emotion_doc if seed_word.has_vector]
                            else:
                                # Use custom vector similarity function
                                similarities = [vector_similarity(ctx.vector, seed_word.vector, metric=similarity_metric) for seed_word in emotion_doc if seed_word.has_vector]
                            if not similarities:
                                continue
                            if similarity_aggregation == "average":
                                avg_similarity = sum(similarities) / len(similarities)
                            elif similarity_aggregation == "max":
                                avg_similarity = max(similarities)

                            #  VAD similarity
                            vad_vec_word = get_vad_vector(ctx.lemma_.lower(), VAD)
                            vad_sim = 0
                            if vad_vec_word is not None and emotion_vad_profiles[emotion] is not None:
                                vad_sim = vad_similarity_alt(vad_vec_word, emotion_vad_profiles[emotion], vad_metric="euclidean")
                            
                            #  Combine
                            final_sim = alpha * avg_similarity + (1 - alpha) * vad_sim
                            emotion_context_scores[emotion].append((final_sim, distance))

        # Aggregate scores (reuse your existing calculate_emotion_scores function)
        all_emotion_context_scores[chap_num].append(emotion_context_scores)
        results[chap_num] = calculate_emotion_scores(emotion_context_scores, aggregation_method, lambda_decay)

    return results, all_emotion_context_scores
```
